```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# GitHub doesn't allow files larger than 100 MB to be uploaded directly to repositories.
# The dataset 'creditcard.csv' is 143 MB, so we use Google Drive to host the file.
# When the script runs, it will download the dataset from Google Drive if it's not already present.

import os
import requests
import pandas as pd

url = 'https://drive.google.com/uc?id=1sTVSN7xMtSTaxGkdJuEwq-0qXxbU6Hg_'

# Check if the file already exists to avoid downloading it multiple times
if not os.path.exists('creditcard.csv'):
    print("Downloading dataset...")
    response = requests.get(url)
    with open('creditcard.csv', 'wb') as f:
        f.write(response.content)
    print("Dataset downloaded successfully.")
else:
    print("Dataset already exists.")
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
# This Dataset is highly unblanced
# 0 --> Normal Transaction
# 1 --> fraudulent transaction

# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
# Under-Sampling
# Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions
# Number of Fraudulent Transactions --> 492

legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the data into Features & Targets

X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
# Split the data into Training data & Testing Data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model Training
# Logistic Regression

model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
# Accuracy Score

# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
